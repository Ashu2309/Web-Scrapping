## a. Fairfax County Public Schools - https://www.fcps.edu/
Using Beautifulsoup

```python
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
import requests
```

```python
no = 0
names = []
links = []
designations = []
while(no<7):
    fcps_url = "https://www.fcps.edu/search?f%5B0%5D=content_type%3Astaff_profile&keywords=staff%20profile&page="+str(no)
    r = requests.get(fcps_url)
    fcps_soup = BeautifulSoup(r.content,'html.parser')
    emp_name = fcps_soup.find_all("div",class_="node__title")
    for name in emp_name:
        i = name.a.text
        d = name.a["href"]
        j = name.next_sibling.next_sibling.text.strip("\n")
        names.append(i)
        links.append("https://www.fcps.edu"+d)
        designations.append(j)
    no = no + 1
```

```python
# print(fcps_soup.prettify())
```

```python
fcps = pd.DataFrame()
fcps["Employee Name"] = names
fcps["Designation"] = designations
fcps["More details"] = links
fcps = fcps.set_index(np.arange(1,len(fcps)+1))
pd.set_option("display.max_rows",None)
```
