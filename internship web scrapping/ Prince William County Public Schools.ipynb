using selenium

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver import ActionChains
from selenium.webdriver.common.keys import Keys
import pandas as pd
import numpy as np
import time
```

```python
driver = webdriver.Chrome(r"C:\Users\Nidhi\Downloads\chromedriver_win32\chromedriver.exe")
driver.maximize_window()
```

```python
driver.get("https://www.pwcs.edu/")
```

```python
all_details = []
names = []
designations = []
mains = []
emails = []

no_of_schools = int(input("How many number school data you want to fetch out off 99 : "))
elems = []
elements = []
```

```python
for key in range(0,no_of_schools):
    if(key == 2):
        continue
    select = driver.find_element(By.XPATH,"//*[contains(text(),'Select a School')]")
    select.click()
    # time.sleep(2)
    
    #clicking on s school
    school_div = driver.find_elements(By.XPATH,'//div[@class="fl w-25"]')
    elems = []
    for i in school_div:
        child = i.find_elements(By.TAG_NAME,"li")
        for j in child:
            a = j.find_elements(By.TAG_NAME,"a")
            for k in a:
                elems.append(k)
#   elements.append(elems[0])
#     print(elems[key])
    elems[key].click()
#     elems.pop(0)
    time.sleep(2)
    staff_dir = driver.find_element(By.XPATH,"//ul[@class='nav-menu']//*[contains(text(),'Staff Directory')]")
    
    action = ActionChains(driver)    
    action.double_click(staff_dir).perform()

    start = 0
    end = driver.find_element(By.XPATH,'//ul[@class="pagination"]/li[last()-2]').text

    for page in range(start,int(end)):
            
        name_det = driver.find_elements(By.XPATH,'//div[@class="panel-title"]')
        designation_det = driver.find_elements(By.XPATH,'//div[@class="text-muted depts"]')
        main_det = driver.find_elements(By.XPATH,'//div[@class="panel-text"]/div[2]')
        email_det = driver.find_elements(By.XPATH,'//div[@class="email"]')

        for name in name_det:
            names.append(name.text)
        for desig in designation_det:
            designations.append(desig.text)
        for main in main_det:
            mains.append(main.text)
        for email in email_det:
            emails.append(email.text)
            
        if(page == int(end)-1):
            break
        else:
            next_button = driver.find_element(By.XPATH,'//ul[@class="pagination"]/li[last()-1]/a')
            next_button.click()
        
```

```python
print("Total no of Schools in data : {} ".format(len(elems)))
```

```python
employee_details = pd.DataFrame({"Employee Name":names,"Designation":designations,"Position":mains,"Email Address":emails})
pd.set_option("display.max_rows",None)
employee_details = employee_details[employee_details["Employee Name"]!=""]
employee_details = employee_details.set_index(np.arange(1,len(employee_details)+1))
```
