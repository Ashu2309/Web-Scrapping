## b. Miami Dade County Public Schools - https://www3.dadeschools.net/home

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
import numpy as np
import time
```

```python
driver = webdriver.Chrome(r"C:\Users\Nidhi\Downloads\chromedriver_win32\chromedriver.exe")
driver.maximize_window()
```

```python
driver.get("https://www3.dadeschools.net/EmployeeSearch/employee?lastName=")
```

```python
all_details = []
names = []
telephones = []
locations = []
location_nos = []
```

The below code changes 10 items per page to 100 items per page.
Note : The data is too big around 36,939 to get all on one page.

```python
time.sleep(3)
select = driver.find_element(By.XPATH,'//select[@id="perpage-select"]')
time.sleep(3)
select.click()
time.sleep(3)
option = driver.find_element(By.XPATH,"//*[contains(text(),'100 items per page')]")
time.sleep(3)
option.click()
time.sleep(3)
```

```python
start = 0
end = int(input("Enter no of pages from which you have to collect data : "))
if(end>370):
    print("No of pages should be in range of (0,370) ")
else:
    for page in range(start,end):
        dade_names = driver.find_elements(By.XPATH,'//td[@class="ng-star-inserted"]')
        # dade_names
        for i in dade_names:
            i = i.text
            all_details.append(i)
        next_button = driver.find_element(By.XPATH,'//a[@aria-label="Next"]')
        next_button.click()
        time.sleep(2)
```
